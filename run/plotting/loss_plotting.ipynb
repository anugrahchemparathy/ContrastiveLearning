```python
import matplotlib.pyplot as plt
```

```python
import os
from pathlib import Path
```

```python
import sys
sys.path.append("../")
```

```python
import numpy as np
```

```python
def plot_loss(model_location, title, save = False):
    loss = np.load(os.path.join(Path(model_location), "loss.npy"))
    #print(loss.shape)
    plt.plot(np.arange(loss.shape[0]), loss)
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    if save:
        plt.savefig(f'{title}.png')
    plt.show()
```

```python
plot_loss("../saved_models/rmse_100_a", "rmseNCE_3D Loss", True)
```
