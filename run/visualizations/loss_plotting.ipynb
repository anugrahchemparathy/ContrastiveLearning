```python
def plot_loss(model_location, title, save = False):
#     loss = np.load(os.path.join(Path(model_location), "loss.npy"))
    loss = return_reshaped_loss(model_location)
    #print(loss.shape)
    plt.plot(np.arange(loss.shape[0]), loss)
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    if save:
        plt.savefig(f'{title}.png')
    plt.show()
```

```python
plot_loss("../saved_models/Cosa", "rmseNCE_3D Loss", True)
```

```python
def return_reshaped_loss(model_location):
    loss = np.load(os.path.join(Path(model_location), "loss.npy"))
    loss_new = np.average(loss.reshape((-1,20)), axis = 1)
    return loss_new
```

```python
loss_a = return_reshaped_loss('../saved_models/L2a')
loss_b = return_reshaped_loss('../saved_models/L2b')
loss_c = return_reshaped_loss('../saved_models/L2c')
loss_d = return_reshaped_loss('../saved_models/L2d')
```

```python
title = 'L2NCE Training Loss over 1500 Epochs for models ultimately converging to Bowl/Disk'
legend = []

plt.figure(figsize=(12, 4))

# for loss,loss_type in zip([loss_a, loss_b, loss_c, loss_d],['Tall Bowl','BN Bowl','Skinny Donut','BN Donut']):
for loss,loss_type in zip([loss_b, loss_d],['Batch Norm Bowl','Batch Norm Disk']):
    plt.plot(np.arange(loss.shape[0]), loss)
    legend.append(loss_type)
   

plt.title(title)
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.legend(legend)

plt.savefig(f'loss.pdf')

plt.show()
```
